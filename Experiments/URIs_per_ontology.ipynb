```python
import requests
import pandas as pd
```

```python
base_url = "http://77.83.99.74:5000/project/metadata/"
```

# Cell Type

```python
response = requests.get(base_url + "cell_type")
results = response.json()
new_results = []

for result in results:
    term = result['term']
    URIs = result['URI']
    
    for URI in URIs:
        if URI is None:
            continue
        
        repo = URI.split('/')[-1].split('_')[0]
        
        new_results.append({
            'term': term,
            'URI': URI,
            'repo': repo
        })
        
cell_types = pd.DataFrame(new_results)
cell_types
```

```python
len(cell_types.term.unique())
```

```python
cell_types_counts = cell_types[['term', 'repo']].groupby(by='repo').count()
cell_types_counts = cell_types_counts.rename(columns={"term": "cell_type_terms"})
cell_types_counts
```

# Disease

```python
response = requests.get(base_url + "disease")
results = response.json()
new_results = []

for result in results:
    term = result['term']
    URIs = result['URI']
    
    for URI in URIs:
        if URI is None:
            continue
        
        repo = URI.split('/')[-1].split('_')[0]
        
        new_results.append({
            'term': term,
            'URI': URI,
            'repo': repo
        })
        
diseases = pd.DataFrame(new_results)
diseases
```

```python
len(diseases.term.unique())
```

```python
diseases_counts = diseases[['term', 'repo']].groupby(by='repo').count()
diseases_counts = diseases_counts.rename(columns={"term": "disease_terms"})
diseases_counts
```

# Organism Part

```python
response = requests.get(base_url + "organism_part")
results = response.json()
new_results = []

for result in results:
    term = result['term']
    URIs = result['URI']
    
    for URI in URIs:
        if URI is None:
            continue
        
        repo = URI.split('/')[-1].split('_')[0]
        
        new_results.append({
            'term': term,
            'URI': URI,
            'repo': repo
        })
        
organism_parts = pd.DataFrame(new_results)
organism_parts
```

```python
len(organism_parts.term.unique())
```

```python
organism_parts_counts = organism_parts[['term', 'repo']].groupby(by='repo').count()
organism_parts_counts = organism_parts_counts.rename(columns={"term": "organism_parts_terms"})
organism_parts_counts
```

```python
aux = pd.merge(cell_types_counts, diseases_counts, how='outer', left_index=True, right_index=True)
repos = pd.merge(aux, organism_parts_counts, how='outer', left_index=True, right_index=True)
repos
```

```python
repos = repos.fillna(0)
repos
```

```python
repos = repos.astype('int32')
repos
```

```python
len(repos)
```

```python
repos['organism_parts_terms'] = repos['organism_parts_terms'] - repos['cell_type_terms']
repos
```

```python
repos.loc['Total'] = repos.sum()
repos
```

```python
repos['total_terms'] = repos.sum(axis=1)
repos
```
